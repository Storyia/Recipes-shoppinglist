```python
#Сбор данных с сайта
import requests            #для HTTP-запросов,для загрузки страниц с сайта
from bs4 import BeautifulSoup                #для парсинга HTML и XML документов.для извлечения нужных данных из HTML кода страницы
import json                                   #для работы с JSON
import pandas as pd                           #для работы с таблицами данных. для создания и сохранения данных в файл CSV

def get_recipes(url):                         #Отправляет HTTP-запрос на указанный URL и получает ответ (страницу)
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')           #Парсит HTML код страницы,чтобы извлечь данные
    script = soup.find('script', string=lambda t: t and 'initialState' in t).text         #Ищет тег <script> со строкой 'initialState',данные рецептов находятся внутри этого тега в виде JSON
    data_start = script.find('{')                       #Определяем начало и конец JSON данных
    data_end = script.rfind('}') + 1
    json_data = json.loads(script[data_start:data_end])              #Преобразуем строку JSON в Python-объект (словарь)
    return json_data['props']['pageProps']['initialState']['recipe']['recipes']

base_url = 'https://eda.ru/recepty'
recipes_data = []            #Пустой список для хранения данных всех рецептов
page = 1                      #первая страница

while True:                #Бесконечный цикл для обхода всех страниц сайта
    current_url = f"{base_url}?page={page}"          #Формируем URL для текущей страницы
    print(f"Загрузка страницы {page}")
    recipes = get_recipes(current_url)               #Получаем рецепты с текущей страницы
    if not recipes:                                  #Проверяем, есть ли рецепты на странице. Если нет, прекращает цикл
        print("Данные на странице отсутствуют. Завершение сбора данных.")
        break
    for recipe in recipes:
        recipes_data.append({                             #Добавляем словарь с информацией о рецепте в список recipes_data  (ввиде строки объединен пробелами)
            'name': recipe['name'],
            'ingredients': " ".join([f"{ingredient['amount']} {ingredient['measureUnit']['name']} {ingredient['ingredient']['name']}" for ingredient in recipe['composition']]),
            'steps': " ".join([step['description'] if step['description'] else "Описание отсутствует" for step in recipe['recipeSteps']])
        })
    page += 1

# Сохраняем в CSV
df = pd.DataFrame(recipes_data)              #Создаем DataFrame из списка рецептов
df.to_csv('all_recipes.csv', index=False)
print("Данные о всех рецептах сохранены в файл 'all_recipes.csv'")
```
